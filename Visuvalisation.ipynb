```python
import numpy as np
import pandas as pd
df1=pd.read_csv('Cleaned Data.csv')
df1.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

columns = ['Quantity','UnitPrice']

for col in columns:
    plt.figure(figsize=(8,6))
    sns.kdeplot(
        x=np.log1p(df[col]),
        fill=True,
        color='skyblue'
    )
    plt.title(f"KDE Plot of {col}")
    plt.xlabel(col)
    plt.ylabel("Density")
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

columns = ['Quantity','UnitPrice','Total']

for col in columns:
    plt.figure(figsize=(8,6))
    sns.kdeplot(
        x=df1[col],
        fill=True,
        color='skyblue'
    )
    plt.title(f"KDE Plot of {col}")
    plt.xlabel(col)
    plt.ylabel("Density")
    plt.show()
```
